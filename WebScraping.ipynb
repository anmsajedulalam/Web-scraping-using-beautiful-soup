**Importing libraries for beautiful soup and url request and os access**

```python
from bs4 import BeautifulSoup
import urllib.request
import os
```

```python
URL = "http://library.bracu.ac.bd/downloads"
```

```python
#req = urllib.request.Request(URL, headers={'User-Agent': 'Mozilla/5.0'})
#page = urllib.request.urlopen(req).read()
page = urllib.request.urlopen(URL)
```

```python
print(page)
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
print(soup)
```

```python
for link in soup.findAll('a'):
  theLink = link.get('href')
  name = link.string

  print(theLink)
  print(name)
```

```python
for link in soup.findAll('a'):
  theLink = link.get('href')
  name = link.string

  if theLink[-4:] == ".doc" or theLink[-4:] == ".pdf":
    if theLink[-4:] == ".doc":
      fileExtension = ".doc"
      if name is None:
        title = "Title is None" + fileExtension
      else:
        title = name + fileExtension
      try:
        with open(os.getcwd() + "/contents/" + title, "r") as fileExists:
          print(title + " already exists")
          fileExists.close()

      except FileNotFoundError:
        print("Starting to download: " + title)
        docFile = open(os.getcwd() + "/contents/" + title, "wb")
        docFile.write(urllib.request.urlopen(theLink).read())
        docFile.close()
        print("Completed.")

    if theLink[-4:] == ".pdf":
      fileExtension = ".pdf"
      if name is None:
        title = "Title is None" + fileExtension
      else:
        title = name + fileExtension
      try:
        with open(os.getcwd() + "/contents/" + title, "r") as fileExists:
          print(title + " already exists")
          fileExists.close()

      except FileNotFoundError:
        print("Starting to download: " + title)
        docFile = open(os.getcwd() + "/contents/" + title, "wb")
        docFile.write(urllib.request.urlopen(theLink).read())
        docFile.close()
        print("Completed.")
```
